```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(grades["value"], bins=20, color="#4F81BD", edgecolor="white",linewidth=0.5 )
ax.axvline(grades["value"].median(), color="#2ECC71", linestyle="--", linewidth=2, label=f"Медіана: {grades["value"].median():.1f}")
ax.set_title("Розподіл оцінок", fontsize=14, fontweight="bold")
ax.set_xlabel("Оцінка")
ax.set_ylabel("Кількість студентів")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
average_grade = grades.groupby("group_id")["value"].mean().reset_index()
average_grade.columns = ["group_id", "avg_grade"]
average_grade = average_grade.sort_values("avg_grade", ascending=False)
print(average_grade)
```

Рейтинг по групах.

```python
student_avg = grades.groupby(["group_id", "student_id"])["value"].mean().reset_index()
student_avg.columns = ["group_id", "student_id", "avg_grade"]

top10 = (
    student_avg
    .sort_values("avg_grade", ascending=False)
    .groupby("group_id")
    .apply(lambda x: x.head(max(1, int(len(x) * 0.4))))
    .reset_index(drop=True)
)

print(top10.to_string(index=False))
```

 Перших 10% найкращих студентів з усіх груп. Рейтинг на стипендію, де колонки: group_id, student_id, average_grade.

```python
student_avg = grades.groupby(["group_id", "student_id"])["value"].mean().reset_index()
student_avg.columns = ["group_id", "student_id", "avg_grade"]
n_top = max(1, int(len(student_avg) * 0.4))
top10 = student_avg.sort_values("avg_grade", ascending=False).head(n_top).copy()
students_users = users[users["role"] == "student"][["id", "full_name"]]
top10 = top10.merge(students_users, left_on="student_id", right_on="id", how="left")
top10 = top10.drop(columns=["id"])
top10["full_name"] = top10["full_name"].fillna("Немає в системі")
print(f"Топ {n_top} студентів з {len(student_avg)} загалом\n")
print(top10[['group_id', 'student_id', 'full_name', 'avg_grade']].to_string(index=False))
```

Рейтинг на стипендію з повними іменами студентів. Немає в системі з'являється через те, що users містить лише 400 студентів, а в grades фігурує 1543 унікальних student_id.
